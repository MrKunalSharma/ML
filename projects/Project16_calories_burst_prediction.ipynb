```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
cs=pd.read_csv('/content/calories.csv')
cs.head()
```

```python
ex=pd.read_csv('/content/exercise.csv')
ex.head()
```

```python
ds=pd.concat([ex,cs.Calories],axis=1)
ds.head()
```

```python
ds.info()
```

```python
ds.replace({"Gender":{"male":0,"female":1}},inplace=True)
```

```python
ds.head()
```

```python
x=ds.drop(columns=['User_ID','Calories'],axis=1)
y=ds.Calories
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_predicition=model.predict(x_train)
print(metrics.mean_absolute_error(x_train_predicition,y_train))
print(metrics.r2_score(x_train_predicition,y_train))
```

```python
x_test_predicition=model.predict(x_test)
print(metrics.mean_absolute_error(x_test_predicition,y_test))
print(metrics.r2_score(x_test_predicition,y_test))
```
