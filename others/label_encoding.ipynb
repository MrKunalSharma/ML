```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
dataset1=pd.read_csv('/content/data.csv')
```

```python
dataset1.head()
```

```python
dataset1.diagnosis.value_counts()
```

```python
labelencode=LabelEncoder()
```

```python
labelencode.fit(dataset1.diagnosis)
```

```python
label1=labelencode.transform(dataset1.diagnosis)
```

```python
dataset1['target']=label1
```

```python
dataset1
```

```python
dataset1.target.value_counts()
```
