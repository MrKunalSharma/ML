**Matrix Addition**

**Matrix Subtraction**

```python
# creating two matrices

A = np.array([[2,3],[4,5]])

B = np.array([[6,7],[8,9]])
```

```python
print(A)
```

```python
print(B)
```

```python
difference = A - B
```

```python
print(difference)
```
