```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```

```python
vector1=np.asarray([0,0,4,5])
vector2=np.asarray([0,0,1,2])
sum=vector1+vector2
print(sum)
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,1,2,scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,5,7,scale_units='xy',angles='xy',scale=1)
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```

```python
vector1=np.asarray([0,0,4,5])
vector2=np.asarray([0,0,1,2])
diff=vector1-vector2
print(diff)
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,1,2,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,3,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```

```python
vector1=np.asarray([0,0,4,5])
mul=2*vector1
print(mul)
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,8,10,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
vector1=np.asarray([0,0,4,5])
mul=0.5*vector1
print(mul)
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,2,2.5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
